```python
import numpy as np
import torch
import torchvision
from torchvision import transforms
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision.utils import save_image
import matplotlib.pyplot as plt
from PIL import Image
import os
import math
```

Datasets Classes

```python
class Div2k(Dataset):
  def __init__(self, root_dir):
    self.root_dir = root_dir
    self.imgs_path = np.array(list(sorted(os.listdir(self.root_dir))))

  def __len__(self):
    return len(self.imgs_path)

  def __getitem__(self, idx):
    if torch.is_tensor(idx):
        idx = idx.tolist()
    hr_in_transform = transforms.Compose([
      transforms.Resize((800, 800)),                                    
      transforms.Resize((224, 224)),
      transforms.ToTensor(),
      transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
    lr_in_transform = transforms.Compose([
      transforms.Resize((800, 800)),                                    
      transforms.Resize((224, 224)),
      transforms.Resize((112, 112)),                                    
      transforms.Resize((224, 224)),
      transforms.ToTensor(),
      transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
    # load image from disk
    img_lr = lr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))
    img_hr = hr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))

    return {"lr": img_lr, "hr": img_hr}
```

```python
class Coco(Dataset):
  def __init__(self, root_dir):
    self.root_dir = root_dir
    self.imgs_path = np.array(list(sorted(os.listdir(self.root_dir))))

  def __len__(self):
    return len(self.imgs_path)

  def __getitem__(self, idx):
    if torch.is_tensor(idx):
        idx = idx.tolist()
    hr_in_transform = transforms.Compose([
      transforms.Resize((480, 480)),                                    
      transforms.Resize((224, 224)),
      transforms.ToTensor(),
      transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
    lr_in_transform = transforms.Compose([
      transforms.Resize((480, 480)),                                    
      transforms.Resize((224, 224)),
      transforms.Resize((112, 112)),                                    
      transforms.Resize((224, 224)),
      transforms.ToTensor(),
      transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
    # load image from disk
    img_lr = lr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))
    img_hr = hr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))

    return {"lr": img_lr, "hr": img_hr}
```

```python
# For datasets that already in 224x224 resolution
class FromFile(Dataset):
  def __init__(self, root_dir):
    self.root_dir = root_dir
    self.imgs_path = np.array(list(sorted(os.listdir(self.root_dir))))

  def __len__(self):
    return len(self.imgs_path)

  def __getitem__(self, idx):
    if torch.is_tensor(idx):
        idx = idx.tolist()
    hr_in_transform = transforms.Compose([
      transforms.ToTensor(),
      transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
    lr_in_transform = transforms.Compose([
      transforms.Resize((112, 112)),                                    
      transforms.Resize((224, 224)),
      transforms.ToTensor(),
      transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
    # load image from disk
    img_lr = lr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))
    img_hr = hr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))

    return {"lr": img_lr, "hr": img_hr}
```

```python
# MSLE need a diffrent normalization
class FromFileMSLE(Dataset):
  def __init__(self, root_dir):
    self.root_dir = root_dir
    self.imgs_path = np.array(list(sorted(os.listdir(self.root_dir))))

  def __len__(self):
    return len(self.imgs_path)

  def __getitem__(self, idx):
    if torch.is_tensor(idx):
        idx = idx.tolist()
    hr_in_transform = transforms.Compose([
      transforms.ToTensor(),
      transforms.Normalize((0, 0, 0), (1, 1, 1))])
    lr_in_transform = transforms.Compose([
      transforms.Resize((112, 112)),                                    
      transforms.Resize((224, 224)),
      transforms.ToTensor(),
      transforms.Normalize((0, 0, 0), (1, 1, 1))])
    # load image from disk
    img_lr = lr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))
    img_hr = hr_in_transform(Image.open(os.path.join(self.root_dir, self.imgs_path[idx])).convert('RGB'))

    return {"lr": img_lr, "hr": img_hr}
```

Additinal Functions

```python
# For saving the dataset in low resolution
def save_low(dataset, path):
    hr_in_transform = transforms.Compose([
      transforms.Resize((800, 800)),                                    
      transforms.Resize((224, 224)),
      transforms.ToTensor()])
    for p in dataset.imgs_path:
      img = hr_in_transform(Image.open(os.path.join(dataset.root_dir, p)).convert('RGB'))
      save_image(img, os.path.join(path, p))
```

```python
def tensor_to_numpy(image_tensor):
  image = image_tensor.to("cpu").clone().detach()
  image = image.numpy().squeeze()
  image = image.transpose(1, 2, 0)
  image = image * np.array((0.229, 0.224, 0.225)) + np.array((0.485, 0.456, 0.406))
  image = image.clip(0, 1)
  
  return image
def MSLE_tensor_to_numpy(image_tensor):
  image = image_tensor.to("cpu").clone().detach()
  image = image.numpy().squeeze()
  image = image.transpose(1, 2, 0)
  image = image * np.array((1, 1, 1)) + np.array((0, 0, 0))
  image = image.clip(0, 1)
  
  return image
```

```python
def PSNR(hr, loss):
  return 10 * math.log10((hr.max() ** 2) / loss)
```

```python
def MSLE(x, y):
  return nn.MSELoss()(torch.log(x + 1), torch.log(y + 1))
def LC(x, y):
  return torch.sum(torch.log(torch.cosh(x - y)))
```

The model

```python
class SRCNN(nn.Module):
    def __init__(self, num_channels=1):
        super(SRCNN, self).__init__()
        self.conv1 = nn.Conv2d(num_channels, 64, kernel_size=9, padding=9 // 2)
        self.conv2 = nn.Conv2d(64, 32, kernel_size=1, padding=1 // 2)
        self.conv3 = nn.Conv2d(32, num_channels, kernel_size=5, padding=5 // 2)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = self.conv3(x)
        return x
```

```python
model = SRCNN(3)
```

Training

```python
def train(model, optimizer, loss_function, train_dataloader, val_loader, 
                             num_epochs=10, print_freq=30, start=0,name="", device="cuda"):
  # move model to device
  model = model.to(device)
  train_losses, val_losses = [], []
  train_psnr, val_psnr = [], []
  mse = nn.MSELoss()
  print("start")
  for epoch in range(num_epochs):
    running_loss = 0
    running_val_loss = 0
    running_psnr = 0
    running_val_psnr = 0
    try:
      model.train()
      loss_value = 0
      for (idx, data) in enumerate(train_dataloader):
        optimizer.zero_grad()
        net_out = model(data["lr"].to(device))
        loss = loss_function(net_out, data["hr"].to(device))
        loss.backward()
        optimizer.step()
        loss_value = loss.cpu().data.item()
        running_loss += loss_value
        running_psnr += PSNR(data["hr"], mse(net_out, data["hr"].to(device)))
      else:
        val_loss = 0
        with torch.no_grad():
          for (idx, data) in enumerate(val_loader):
            net_out = model(data["lr"].to(device))
            loss = loss_function(net_out, data["hr"].to(device))
            running_val_loss += loss.item()
            running_val_psnr += PSNR(data["hr"], mse(net_out, data["hr"].to(device)))
      running_psnr /= len(train_dataloader)
      running_val_psnr /= len(val_loader)
      train_losses.append(running_loss)
      val_losses.append(running_val_loss)
      train_psnr.append(running_psnr)
      val_psnr.append(running_val_psnr)
      if epoch % print_freq == 0:
        print(f"Epoch {epoch + start}, Train loss={running_loss}, Train Averaged PSNR={running_psnr}, Val loss={running_val_loss}, Val Averaged PSNR={running_val_psnr}")
      torch.save(model.state_dict(), "/content/model" + name + str(epoch + start))
      f = open("values" + name + str(epoch + start) + ".txt", "w")
      f.write("train_losses: " + str(train_losses) + "\n")
      f.write("val_losses: " + str(val_losses) + "\n")
      f.write("train_psnr: " + str(train_psnr) + "\n")
      f.write("val_psnr: " + str(val_psnr) + "\n")
      f.close()
    except KeyboardInterrupt:
        print(f"Epoch {epoch + start}, Train loss={running_loss}, Train Averaged PSNR={running_psnr}, Val loss={running_val_loss}, Val Averaged PSNR={running_val_psnr}")
        return model, train_losses, val_losses, train_psnr, val_psnr

  return model, train_losses, val_losses, train_psnr, val_psnr
```

```python
optimizer = optim.Adam([
    {'params': model.conv1.parameters()},
    {'params': model.conv2.parameters()},
    {'params': model.conv3.parameters(), 'lr': 0.001 * 0.1}
  ], lr=0.001)
#loss_f = MSLE
#loss_f = LC
loss_f = nn.MSELoss()
#loss_f = nn.L1Loss()
```

```python
model, train_losses, val_losses, train_psnr, val_psnr = train(model, optimizer, loss_f, train_loader, valid_loader, num_epochs=100, print_freq=1, start=0)
```

Validating

```python
def val(model, val_loader, val_dataset, device="cuda"):
  model = model.to(device)
  #model.eval()
  running_val_psnr = 0
  mse = nn.MSELoss()
  for (idx, data) in enumerate(val_loader):
    net_out = model(data["lr"].to(device))
    running_val_psnr += PSNR(data["hr"], mse(net_out, data["hr"].to(device)))
  return running_val_psnr / (len(val_dataset) / 64)
```
